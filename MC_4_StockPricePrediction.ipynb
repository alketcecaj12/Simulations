# MC for stock price prediction

an example of using Monte Carlo simulations for stock price prediction using Python. This example will demonstrate how to simulate future stock prices for a given company.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch historical data for a stock (e.g., Apple)
ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2023-12-31"
stock_data = yf.download(ticker, start=start_date, end=end_date)
```

# K parameters needed for simulation

```python
# Calculate daily returns
stock_data['Returns'] = stock_data['Adj Close'].pct_change()

# Calculate annualized drift and volatility
drift = stock_data['Returns'].mean()
volatility = stock_data['Returns'].std()

# Number of simulations and time steps
num_simulations = 1000
num_days = 252  # Typical number of trading days in a year
```

### Let's create a function to perform the Monte Carlo simulation

```python
def monte_carlo_simulation(start_price, drift, volatility, days, simulations):
    dt = 1 / days
    simulated_prices = np.zeros((days, simulations))
    simulated_prices[0] = start_price
    
    for i in range(1, days):
        Z = np.random.standard_normal(simulations)
        simulated_prices[i] = simulated_prices[i-1] * np.exp((drift - 0.5 * volatility**2) * dt + volatility * np.sqrt(dt) * Z)
    
    return simulated_prices

# Run the simulation
start_price = stock_data['Adj Close'].iloc[-1]
simulated_prices = monte_carlo_simulation(start_price, drift, volatility, num_days, num_simulations)
```

### Visualize the results

```python
plt.figure(figsize=(10, 6))
plt.plot(simulated_prices)
plt.title(f"Monte Carlo Simulation: {ticker} Stock Price")
plt.xlabel("Trading Days")
plt.ylabel("Stock Price")
plt.show()
```

# Analyzing the simulation results

```python
# Calculate the expected price and confidence intervals
expected_price = np.mean(simulated_prices[-1])
confidence_interval = np.percentile(simulated_prices[-1], [2.5, 97.5])

print(f"Expected stock price after {num_days} trading days: ${expected_price:.2f}")
print(f"95% Confidence Interval: ${confidence_interval[0]:.2f} - ${confidence_interval[1]:.2f}")

# Plot a histogram of final prices
plt.figure(figsize=(10, 6))
plt.hist(simulated_prices[-1], bins=50)
plt.title(f"Distribution of Simulated {ticker} Stock Prices")
plt.xlabel("Stock Price")
plt.ylabel("Frequency")
plt.axvline(expected_price, color='r', linestyle='dashed', linewidth=2, label='Expected Price')
plt.legend()
plt.show()
```

# Explanation of the model

This Monte Carlo simulation example for stock price prediction does the following:
Fetches historical stock data for Apple (AAPL) using the yfinance library. 

 - Calculates daily returns, drift, and volatility based on historical data2.
 - Performs 1000 simulations of stock prices for the next 252 trading days (approximately one year)3.
 - Visualizes the simulated stock price paths4.
 - Calculates and displays the expected stock price and a 95% confidence interval5.

This simulation provides insights into the potential future stock prices based on historical volatility and returns.
